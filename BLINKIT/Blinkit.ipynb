```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_excel("BlinkIT Grocery Data.xlsx")
df=pd.DataFrame(data)
print(df.head(10))
```

```python
df.isnull().sum()
```

```python
df.select_dtypes(include='number').corr()
```

```python
top_sales=df.groupby(['Item Type','Outlet Location Type','Outlet Type'])['Sales'].max().nlargest(10)
print(top_sales)
```

```python
top_avg_sales=df.groupby(['Item Type','Outlet Location Type','Outlet Type'])['Sales'].mean().nlargest(10)
print(top_avg_sales)
```

```python
best_rating=df.groupby(['Item Type','Outlet Location Type','Outlet Type'])['Rating'].max().nlargest(10)
print(best_rating)
```

```python
best_avg_rating=df.groupby(['Item Type','Outlet Location Type','Outlet Type'])['Rating'].mean().nlargest(10)
print(best_avg_rating)
```

```python
best_product= df.groupby(['Item Type','Outlet Location Type','Outlet Type'])['Sales'].sum().nlargest(10)
print(best_product)
```

```python
sns.relplot(data=df,x="Rating",y="Sales")
plt.show()
```

```python
plt.title("Rating vs Item Type")
sns.barplot(data=df,x="Item Type",y="Rating")
plt.xticks(rotation=90)
plt.xlabel("Item Type")
plt.ylabel("Rating")

plt.show()

```

```python
sns.violinplot(data=df,x="Sales")
plt.title("Sales Distribution")
plt.xlabel("Sales")

plt.show()
```

```python
sns.violinplot(data=df,x="Rating")
plt.title("Rating Distribution")
plt.xlabel("Rating")

plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
